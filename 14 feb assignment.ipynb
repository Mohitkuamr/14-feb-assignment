```python
Ans = 4
import threading

def squares():
    for i in range(1, 6):
        print(f"{i} squared is {i*i}")

def cubes():
    for i in range(1, 6):
        print(f"{i} cubed is {i*i*i}")

t1 = threading.Thread(target=squares)
t2 = threading.Thread(target=cubes)

t1.start()
t2.start()

t1.join()
t2.join()

print("Main thread completed")
```

```python
Ans = 5
Advantages:

1.
Improved performance: Multithreading can improve the performance of a program by allowing it to perform multiple tasks concurrently.
This can help to reduce the overall execution time of the program.

2.
Better resource utilization: Multithreading allows a program to utilize system resources more efficiently by running multiple tasks 
concurrently. This can help to reduce the overall resource consumption of the program.

3.
Simplified code: Multithreading can simplify the code of a program by allowing it to perform multiple tasks concurrently. 
This can help to reduce the complexity of the program and make it easier to maintain.

4.
Responsiveness: Multithreading can make a program more responsive by allowing it to perform multiple tasks concurrently.
This can help to improve the user experience and make the program more user-friendly.

Disadvantages:

1. 
Increased complexity: Multithreading can increase the complexity of a program by introducing new synchronization and coordination 
challenges. This can make the program harder to develop and debug.

2. 
Synchronization issues: Multithreading can introduce synchronization issues, such as race conditions and deadlocks, 
which can be difficult to detect and resolve.

3.
Resource contention: Multithreading can result in resource contention, where multiple threads compete for the same resource,
such as a shared variable or a file. This can lead to performance degradation and other issues.

4
Overhead: Multithreading can introduce overhead, such as the overhead associated with context switching and synchronization.
This can reduce the overall performance of the program.
```
